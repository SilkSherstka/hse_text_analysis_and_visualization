```python
from google.colab import files
import pandas as pd

# Загружаем файл с компьютера
uploaded = files.upload()

# Получаем имя файла
file_name = list(uploaded.keys())[0]

# Читаем Excel файл в DataFrame
df = pd.read_excel(file_name)

# Показать первые строки
print(df.head())
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Создаем новую колонку с длиной текста
df['text_length'] = df['text'].apply(len)

#Cначала я попробовала построить такую визуализацию с помощью boxplot,
#но через boxplot визуализацию было сложно читать. Через scatter в итоге получилось более наглядно

#plt.figure(figsize =(7,5))
#sns.boxplot(data=df, x=df['emotion'], y=df['text_length'])
#plt.title('Длина текста в разрезе эмоций')
#plt.xlabel('Эмоция')
#plt.ylabel('Длина текста (в символах)')
#plt.show()

plt.rcParams['figure.dpi'] = 120
sns.set_theme(style='whitegrid')

plt.figure(figsize = (6,5))
plt.scatter(df['text_length'], df['emotion'], alpha=0.55)
plt.title('Зависимость эмоции от длины текста')
plt.xlabel('Длина текста (символы)')
plt.ylabel('Эмоция')
plt.show()
```

```python
# Подсчитываем количество каждой эмоции
emotion_counts = df['emotion'].value_counts()

# Создаем bar plot
plt.figure(figsize=(12, 6))
bars = plt.bar(emotion_counts.index, emotion_counts.values)

# Добавляем значения на столбцы
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.5,
             f'{int(height)}', ha='center', va='bottom')

plt.title('Распределение эмоций в датасете')
plt.xlabel('Эмоция')
plt.ylabel('Количество записей')
plt.show()
```
